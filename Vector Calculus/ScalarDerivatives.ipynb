```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from PIL import Image
```

### The power rule

```python
x = np.linspace(-10, 10, 400)

y2 = 5 * x**4
y1 = x ** 5
```

```python
plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='y = 2x', color='blue')

plt.plot(x, y2, label='y = x^2', color='red')

plt.title('Plot of y = 5(x^4) and y = x^5')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.show()
```

### The sum rule

```python
x = np.linspace(0, 2 * np.pi, 1000)
f1 = (x**2) + 45
f2 = x**3
df1 = 2*x
df2 = 3 * (x**2)
sum_f = f1 + f2
sum_df = df1 + df2
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, f1, label="f1(x) = sin(x)")
plt.plot(x, f2, label="f2(x) = cos(x)")
plt.plot(x, sum_f, label="f1(x) + f2(x)", linestyle="--")
plt.title("Functions and Their Sum")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x, df1, label="f1'(x) = cos(x)")
plt.plot(x, df2, label="f2'(x) = -sin(x)")
plt.plot(x, sum_df, label="f1'(x) + f2'(x)", linestyle="--")
plt.title("Derivatives and Their Sum")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

### The product rule

```python
x = np.linspace(0, 5, 1000)
f = x**2
g = np.exp(x)
df = 2 * x
dg = np.exp(x)
product = f * g
product_derivative = df * g + f * dg
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(x, f, label="f(x) = x^2")
plt.plot(x, g, label="g(x) = e^x")
plt.plot(x, product, label="f(x)g(x)", linestyle="--")
plt.title("Functions and Their Product")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x, df, label="f'(x) = 2x")
plt.plot(x, dg, label="g'(x) = e^x")
plt.plot(x, product_derivative, label="[f(x)g(x)]' = f'(x)g(x) + f(x)g'(x)", linestyle="--")
plt.title("Derivatives and the Product Rule")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

### The quotient rule

```python
x = np.linspace(0.1, 5, 1000)
f = x**2
g = np.exp(x)
df = 2 * x
dg = np.exp(x)
quotient = f / g
quotient_derivative = (df * g - f * dg) / g**2
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(x, f, label="f(x) = x^2")
plt.plot(x, g, label="g(x) = e^x")
plt.plot(x, quotient, label="f(x)/g(x)", linestyle="--")
plt.title("Functions and Their Quotient")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x, df, label="f'(x) = 2x")
plt.plot(x, dg, label="g'(x) = e^x")
plt.plot(x, quotient_derivative, label="[f(x)/g(x)]' = (f'(x)g(x) - f(x)g'(x))/g(x)^2", linestyle="--")
plt.title("Derivatives and the Quotient Rule")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

### The chain rule

```python
x = np.linspace(0, 5, 1000)
f = np.sin(x)
g = x**2
f_prime = np.cos(x)
g_prime = 2 * x
chain_rule_result = f_prime * g_prime
composite_function = np.sin(x**2)
```

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(x, f, label="f(x) = sin(x)")
plt.plot(x, g, label="g(x) = x^2")
plt.plot(x, composite_function, label="f(g(x)) = sin(x^2)", linestyle="--")
plt.title("Functions and Their Composition")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x, f_prime, label="f'(x) = cos(x)")
plt.plot(x, g_prime, label="g'(x) = 2x")
plt.plot(x, chain_rule_result, label="f'(g(x)) * g'(x)", linestyle="--")
plt.title("Derivatives and the Chain Rule")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

### Trigo and calculus

```python
x = np.linspace(0, 2 * np.pi, 1000)
time = np.linspace(0, 2 * np.pi, 100)
```

```python
functions = {
    "cos(x)": lambda x, t: np.cos(x - t),
    "sin(x)": lambda x, t: np.sin(x - t),
    "-sin(x)": lambda x, t: -np.sin(x - t),
    "-cos(x)": lambda x, t: -np.cos(x - t),
}
```

```python
fig, ax = plt.subplots()
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1.5, 1.5)
ax.set_title("Oscillations of cos(x) and sin(x) with their derivatives")
ax.set_xlabel("x")
ax.set_ylabel("y")

lines = {label: ax.plot([], [], label=label)[0] for label in functions.keys()}
ax.legend(loc="upper right")
```

```python
def init():
    for line in lines.values():
        line.set_data([], [])
    return lines.values()
```

```python
def update(frame):
    t = time[frame]
    for label, func in functions.items():
        y = func(x, t)
        lines[label].set_data(x, y)
    return lines.values()
```

```python
ani = FuncAnimation(fig, update, frames=len(time), init_func=init, blit=True)

ani.save("oscillations_derivatives.gif", writer="pillow", fps=30)
plt.close()

print("GIF saved as 'oscillations_derivatives.gif'")
```
